Grouping my data by generation and creating a graph of suicide rates separated by generation

```python
#This groups the dataset based on the values in the generation column (separates into buckets)
#agg (aggregate) then sums up the values in the named column (suicides_no and population) 

gens = suicide_data.groupby('generation').agg(
    sui_num=('suicides_no', 'sum'),
    pop=('population', 'sum')
)

# Creates a new column (suicides_per_100k) and calculates it 

gens['suicides_per_100k'] = (gens['sui_num'] / gens['pop'])*100000

#This sorts the result by the suicide rate in decending order

gens = gens.sort_values(by='suicides_per_100k', ascending=False)

# Outputs the data

gens
```

```python
gens.plot( y='suicides_per_100k', kind='bar', color='pink')
```

Grouping my data by age and creating a visualization showing suicide rate by age group

```python
ages = suicide_data.groupby('age').agg(
    sui_num=('suicides_no', 'sum'),
    pop=('population', 'sum')
).reset_index()

ages['suicides_per_100k'] = (ages['sui_num'] / ages['pop'])*100000

ages = ages.sort_values(by='suicides_per_100k', ascending=False)

ages
```

```python
ages.plot(y='suicides_per_100k', kind='bar', color='pink')
```

Trying a different approach to understanding grouby 

```python
age_year = suicide_data.groupby(['year', 'age']).agg(
    sui_num=('suicides_no', 'sum'),
    pop=('population', 'sum')
).reset_index()

age_year['suicides_per_100k'] = (age_year['sui_num'] / age_year['pop'])*100000

pivot_age_year = age_year.pivot(index = 'year', columns='age', values='suicides_per_100k')

pivot_age_year
```

```python
plt.figure(figsize=(10,6))

for age_group in pivot_age_year.columns:
    plt.plot(pivot_age_year.index, pivot_age_year[age_group], label=age_group)

plt.title('Suicide Rate per 100k Population by Age Group Over Time', fontsize=14)
plt.xlabel('Year', fontsize=10)
plt.ylabel('Suicide Rate per 100k', fontsize=10)
plt.legend(title='Age Group', fontsize=10)

plt.show()
```

```python
import plotly.graph_objects as go


fig = go.Figure()

fig.add_trace(go.Scatter(x=pivot_gender_year.index, y=pivot_gender_year['female'],
                         mode='lines+markers',
                         name='Female',
                         line=dict(color='#D9A1A7')))

fig.add_trace(go.Scatter(x=pivot_gender_year.index, y=pivot_gender_year['male'],
                         mode='lines+markers',
                         name='Male',
                         line=dict(color='#A3CADB')))

fig.update_layout(
    title='Suicide Rates by Gender Over Time',
    xaxis_title='Year',
    yaxis_title='Suicides per 100k Population',
    legend_title='Sex',
    hovermode='x unified',  
    template='plotly_white'  
)

fig.show()
```

```python
gender_year
```

```python
years = gender_year['year'].unique()

chosen_colors = ['#D9A1A7', '#A3CADB']

fig = go.Figure()

for year in years:
    filtered_df = gender_year[gender_year['year'] == year]
    
    fig.add_trace(go.Pie(
        labels=filtered_df['sex'],
        values=filtered_df['sui_num'],
        name=f'Suicide Rate for {year}',
        domain={'x': [0, 1], 'y': [0, 1]},  
        marker=dict(colors=chosen_colors),
        visible=False  
    ))


fig.data[0].visible = True

dropdown_buttons = []
for i, year in enumerate(years):
    
    dropdown_buttons.append(
        dict(
            label=str(year),
            method="update",
            args=[{"visible": [i == j for j in range(len(years))]},  
                  {"title": f"Suicide Numbers by Gender for {year}"}] 
        )
    )

fig.update_layout(
    updatemenus=[{
        'buttons': dropdown_buttons,
        'direction': 'down',
        'showactive': True,
    }],
    title=f'Suicide Numbers by Gender for {years[0]}',  
    template='plotly_white'
)

fig.show()
```
